### To compare summary output with fmg:
##### produce 1) the sum of n_mapped_reads across each *.summary.csv file in /outputs/leftover/ 

echo "sample_id,total_mapped_reads" > mapped_reads_summary.csv
for f in *.summary.csv; do
    sample=$(awk -F',' 'NR==2{print $12}' "$f")
    sum=$(awk -F',' 'NR>1{sum+=$13} END{print sum}' "$f")
    echo "$sample,$sum" >> mapped_reads_summary.csv
done

#### produce 2) the total number of reads in outputs/:

##### produce {sample}.info.yaml in outputs/ by `summarize_sample_info`
genome-grist run conf.yml summarize_sample_info -c 100 -j 100

```python
### Extract two info to csv file in 

echo "sample_id,n_reads" > total_reads_summary.csv
for f in *.info.yaml; do
    awk '
        /^sample:/   {sample=$2}
        /^n_reads:/  {reads=$2}
        END          {print sample "," reads}
    ' "$f" >> total_reads_summary.csv
done
```

```python
# scp two files to local drive, compute f_mapped_reads = total_mapped_reads/total_reads

import pandas as pd

# Read input files
mapped_df = pd.read_csv("mapped_reads_summary.csv")
total_df = pd.read_csv("total_reads_summary.csv")

# Inspect
mapped_df.head(), total_df.head()
```

```python
merged = pd.merge(
    mapped_df,
    total_df,
    on="sample_id",
    how="inner"
)

merged.head()
```

```python
merged["f_mapped_reads"] = (
    merged["total_mapped_reads"] / merged["n_reads"]
)

merged.head()
```

```python
f_mapped_df = merged[
    ["sample_id", "total_mapped_reads", "n_reads", "f_mapped_reads"]
]
```

```python
f_mapped_df.to_csv("f_mapped_reads.csv", index=False)
```

```python
import pandas as pd

# Load CSVs
fmg = pd.read_csv("../../analysis_g_k21/join_g_k21.csv")
grist = pd.read_csv("f_mapped_reads.csv")

# Inner join on accession/sample_id
joined = fmg.merge(
   grist,
    left_on="accession",
    right_on="sample_id",
    how="inner"
)

# (Optional) drop duplicate ID column
joined = joined.drop(columns=["sample_id"])

# Save result
joined.to_csv("join_g_k21_with_f_mapped_reads.csv", index=False)

len(joined)
```

```python
print(joined.head())
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7, 6))

for biome, sub in joined.groupby("biome3"):
    plt.scatter(
        sub["ref_f_weighted"],
        sub["f_mapped_reads"],
        alpha=0.6,
        s=20,
        label=biome
    )

plt.xlabel("ref_f_weighted")
plt.ylabel("f_mapped_reads")
plt.title("ref_f_weighted vs f_mapped_reads (by biome3)")

plt.legend(
    bbox_to_anchor=(1.05, 1),
    loc="upper left",
    fontsize=7
)

plt.tight_layout()
plt.savefig("ref_f_weighted_vs_f_mapped_reads_biome3.png", dpi=300)
plt.show()
```

```python
# add a “shadow” that shows the overall trend (without cluttering the biome colors)
# overlay a smoothed trend band behind the scatter points.
# LOWESS trend with shaded band (add a smooth curve + translucent band behind points)

import numpy as np
import matplotlib.pyplot as plt
from statsmodels.nonparametric.smoothers_lowess import lowess

x = joined["ref_f_weighted"].values
y = joined["f_mapped_reads"].values

# LOWESS smoothing
smoothed = lowess(y, x, frac=0.3, return_sorted=True)
xs, ys = smoothed[:, 0], smoothed[:, 1]

plt.figure(figsize=(6, 6))

# --- SHADOW (trend band) ---
plt.fill_between(
    xs,
    ys - 0.05,
    ys + 0.05,
    alpha=0.2,
    zorder=1
)

# --- TREND LINE ---
plt.plot(xs, ys, linewidth=2, zorder=2)

# --- SCATTER (by biome3) ---
for biome, sub in joined.groupby("biome3"):
    plt.scatter(
        sub["ref_f_weighted"],
        sub["f_mapped_reads"],
        s=25,
        alpha=0.7,
        zorder=3
    )

plt.xlabel("ref_f_weighted")
plt.ylabel("f_mapped_reads")
plt.title("ref_f_weighted vs f_mapped_reads\nLOWESS trend shadow")

plt.tight_layout()
plt.savefig("ref_f_weighted_vs_f_mapped_reads_shadow.png", dpi=300)
plt.show()
```

```python
# global Linear regression + confidence band (simpler, stricter)

import numpy as np

x = joined["ref_f_weighted"].values
y = joined["f_mapped_reads"].values

# Linear fit
coef = np.polyfit(x, y, 1)
poly = np.poly1d(coef)

xs = np.linspace(x.min(), x.max(), 200)
ys = poly(xs)

plt.figure(figsize=(6, 6))

# Shadow band
plt.fill_between(
    xs,
    ys - 0.05,
    ys + 0.05,
    alpha=0.2,
    zorder=1
)

# Trend line
plt.plot(xs, ys, linewidth=2, zorder=2)

# Scatter
for biome, sub in joined.groupby("biome3"):
    plt.scatter(
        sub["ref_f_weighted"],
        sub["f_mapped_reads"],
        s=25,
        alpha=0.7,
        zorder=3
    )

plt.xlabel("ref_f_weighted")
plt.ylabel("f_mapped_reads")
plt.title("ref_f_weighted vs f_mapped_reads\nLinear trend shadow")

plt.tight_layout()
plt.show()
```

```python
# add three biome-specific legend and shadow trends
# Ensure biome1 ordering & colors (explicit, stable)
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.nonparametric.smoothers_lowess import lowess

# Explicit order
biome_order = ["Host-associated", "Engineered", "Environmental"]

# Explicit colors (use matplotlib defaults or tweak)
biome_color = {
    "Host-associated": "#1f77b4",
    "Engineered": "#ff7f0e",
    "Environmental": "#2ca02c",
}
```

```python
# Scatter plot (unchanged logic, ordered)
plt.figure(figsize=(6, 6))

for biome1 in biome_order:
    sub = joined[joined["biome1"] == biome1]
    plt.scatter(
        sub["ref_f_weighted"],
        sub["f_mapped_reads"],
        s=25,
        alpha=0.7,
        color=biome_color[biome1],
        label=biome1,
        zorder=3
    )
```

```python
# Shadow + trend per biome1, each biome gets its own LOWESS shadow, colored to match
for biome1 in biome_order:
    sub = joined[joined["biome1"] == biome1]

    x = sub["ref_f_weighted"].values
    y = sub["f_mapped_reads"].values

    if len(x) < 5:
        continue  # skip tiny groups safely

    smoothed = lowess(y, x, frac=0.4, return_sorted=True)
    xs, ys = smoothed[:, 0], smoothed[:, 1]

    # Shadow band
    plt.fill_between(
        xs,
        ys - 0.05,
        ys + 0.05,
        color=biome_color[biome1],
        alpha=0.15,
        zorder=1
    )

    # Trend line
    plt.plot(
        xs,
        ys,
        color=biome_color[biome1],
        linewidth=2,
        zorder=2
    )
```
